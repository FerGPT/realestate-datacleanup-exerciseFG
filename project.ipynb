#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

Hint: You can help yourself by reviewing the *multihist* demo of Matplotlib.

```python
# TODO
import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv('assets/real_estate.csv', sep=';')

population_filter = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

filtered_data = ds[ds['level5'].isin(population_filter)]

# Normalizaciñón del precio para las diferentes poblaciones:

filtered_data['normalized_price'] = filtered_data.groupby('level5')['price'].transform(
    lambda x: (x - x.min()) / (x.max() - x.min())
)

# Gráfica
plt.figure(figsize=(12, 8))

# Histogramas para cada población:
for city in population_filter:
    plt.hist(filtered_data[filtered_data['level5'] == city]['normalized_price'], 
             bins=10, 
             alpha=0.5, 
             label=city)

# Labesls
plt.title('Precio Normalizado por población')
plt.xlabel('Precio Normalizado')
plt.ylabel('Frecuencia')
plt.legend(title='Poblaciones')
plt.grid(axis='y')

plt.show()
```
